# Web Scraping

```python
import requests
from bs4 import BeautifulSoup

url = "https://fbref.com/en/players/aa/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table containing player data
    player_table = soup.find("div", class_="section_content")
    # print(player_table)
    if player_table:
        # Find all rows in the table
        rows = player_table.find_all("p")
        links = []
        # Iterate over each row and extract data
        for row in rows:
            # Find all cells in the row
            href = row.find('a')['href']
            links.append(href)
        print(links)
    else:
        print("Player table not found on the page.")
else:
    print("Failed to retrieve page:", response.status_code)


```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

numerical_cols = ['MP', 'Starts', 'Min', '90s', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR']
numerical_cols = ['Tkl', 'Blocks', 'Sh', 'Clr', 'SCA90', 'Int', 'Ast', 'xA', 'CrsPA', 'PrgP', 'PrgC', 'Gls', 'Sh', 'xG']

# Standardize the numerical columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(df[numerical_cols])
print(cosine_sim)
# Function to get similar players
def get_similar_players(player_index, cosine_sim, df):
    # Get similarity scores of the player with others
    sim_scores = list(enumerate(cosine_sim[player_index]))
    # Sort players based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get top 5 similar players (excluding the player itself)
    similar_players = sim_scores[1:6]
    # Get player names
    player_indices = [i[0] for i in similar_players]
    similar_player_names = df['Player'].iloc[player_indices]
    return similar_player_names

# Test with example player
player_name = "Harry Kane"
player_index = df.index[df['Player'] == player_name].tolist()[0]
similar_players = get_similar_players(player_index, cosine_sim, df)
print(f"Players similar to {player_name}: {similar_players.tolist()}")


# player_name = "Harry Kane"
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

def find_similar_players(input_values, df, selected_columns):
    # Normalize the input values
    input_values = pd.DataFrame([input_values], columns=selected_columns)
    scaler = MinMaxScaler()
    input_values[selected_columns] = scaler.fit_transform(input_values[selected_columns])

    # Normalize the selected columns in the dataframe
    df_selected = df[selected_columns]
    df_selected = pd.DataFrame(scaler.fit_transform(df_selected), columns=selected_columns)

    # Calculate cosine similarity between input values and dataframe
    similarity_scores = cosine_similarity(input_values, df_selected)

    # Sort players based on similarity scores
    similar_players_indices = similarity_scores.argsort()[0][::-1]
    similar_players = df.iloc[similar_players_indices]

    return similar_players

# Sample usage
data = {
    "Player": ["Alisson", "Mohamed Salah", "Virgil van Dijk", "Dominik Szoboszlai", "Trent Alexander-Arnold", "Luis Díaz", "Alexis Mac Allister"],
    "Nation": ["br", "eg", "nl", "hu", "eng", "co", "ar"],
    "Pos": ["GK", "FW", "DF", "MF", "DF", "FW", "MF"],
    "Age": [31.124, 31.233, 32.210, 23.101, 25.119, 27.021, 25.041],
    "Gls": [0, 14, 1, 3, 2, 4, 1],
    "Ast": [0, 8, 2, 2, 3, 2, 1],
    "xG": [0, 14.3, 1.6, 2.1, 1.8, 4.4, 0.3],
    "npxG": [0, 9.4, 1.6, 2.1, 1.5, 4.4, 0.3],
    "xAG": [0, 8.5, 1.1, 4.2, 5.1, 1.7, 0.6],
}

df = pd.DataFrame(data)

selected_columns = ["Gls", "Ast", "xG", "npxG", "xAG"]

input_values = [0.5, 0.2, 0.4, 0.3, 0.1]  # Example input values for selected columns

similar_players = find_similar_players(input_values, df, selected_columns)
print(similar_players)
```
